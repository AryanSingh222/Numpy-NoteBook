```python
import numpy as np
import time
```

## Creating Array

```python
arr_1d = np.array([1, 2, 3 , 4 , 5 , 6 , 7])
print(arr_1d)

arr_2d = np.array([[1 , 2 , 3], [4 , 5 , 6]])
print(arr_2d)
```

## List vs NumPy array

```python
py_list = [1, 2 , 3]
print("Python list multiplication : ",py_list)

np_array = np.array([1 , 2 , 3]) ## Element wise multiplication 
print("Python array multiplication : ", np_array * 2)

start_time = time.time()
py_list = [i * 2 for i in range(10000)]
print("\n List operation time : ", time.time() - start_time)

start_time = time.time()
np_array = np.arange(100000)
print("\n Numpy operation time : ", time.time() - start_time)
```

## Creating Array from Scratch

```python
zeros = np.zeros((3,4))
print("Zeros Array : \n",zeros)

ones = np.ones((2,3))
print("Ones Array : \n",ones)

full = np.full((2,2),5) ## Constant Variable
print("full Array : \n",full)

## Random values array 

randoms = np.random.random((2,3)) ## 2 by 3 matrices with Random values
print("Random Array : \n",randoms)

## Sequence Array

sequence = np.arange(0,20,2)
print("Sequence Array : \n",sequence)
```

## Vector , Matrices , Tensor

```python
vector = np.array((1,2,3))
print("Vector :",vector)

matrices = np.array([[1 , 2 , 3],
                     [4 , 5 , 6]])
print("Matrices :",matrices)

tensor = np.array([[[1 , 2] , [3 , 4]],
                   [[5 , 6] , [7 , 8]]])
print("Tensor :",tensor)
```

# Array Properties

```python
arr = np.array([[1 , 2 , 3],
                [4 , 5 , 6]])
print("Shape : ", arr.shape) # Array Shape
print("Dimension : ", arr.ndim) # Array Dimension
print("Size : ", arr.size) # Size of Array
print("Data type : ",arr.dtype) # Data Type
```

# ReShaped Array

```python
arr = np.arange(12)
print("Original array : ",arr)

reshaped = arr.reshape((3,4))
print("\n Reshaped array : \n", reshaped)

# faltten ( returns a new copy of the original array collapsed into one dimension)
flattened = reshaped.flatten()
print("\n Flattened array : ",flattened)

# ravel (return view, instead of copy)
raveled = reshaped.ravel()
print("\n Raveled Array : ",raveled)

# Transpose

transpose = reshaped.T
print("\n Transposed array : ", transpose)
```
